```python
## UNIFY SAME VIDEOS
def unifyVideos(df):
    ## CREATE A NEW DATAFRAME
    df_cols = [
        'video_name',
        'video_channel_name', 
        'video_url', 
        'rank_trend', 
        'likes', 
        'views', 
        'comments', 
        'trending_start_date', 
        'trending_end_date', 
        'video_channel_account', 
        'keywords'
    ]

    ## GET UNIQUE ARRAY
    video_list = df['video_url'].unique()
    ## CREATE FINAL DATAFRAME
    df_final = pd.DataFrame(columns=df_cols)
    for url in video_list:
        print('URL: ' + url)
        dicionario = {}
        ## FILTER URLS TO CHECK DUPLICATIONS
        df1 = df[df['video_url'] == url].sort_values(by='ranking_date', ascending=True)
        ## IF HAVE DUPLICATES UNIFY THE VALUES
        if df1['video_url'].count() > 1:
            ## INCLUDING VALUES INTO NEW DATAFRAME
            dicionario['video_name'] = df1['video_name'].iloc[0]
            dicionario['video_channel_name'] = df1['video_channel_name'].iloc[0]
            dicionario['video_url'] = df1['video_url'].iloc[0]
            dicionario['video_channel_account'] = df1['video_channel_account'].iloc[0]
            dicionario['keywords'] = df1['keywords'].iloc[0]
            dicionario['rank_trend'] = df1['rank_trend'].to_numpy()
            
            ## GET LAST VIDEO OBSERVATION (UPDATED VALUES ABOUT VIDEO)
            obs_last_reg = df1[df1['ranking_date'] == df1['ranking_date'].max()]
            ## GET FIRST VIDEO OBSERVATION (UPDATED VALUES ABOUT VIDEO)
            obs_fst_reg = df1[df1['ranking_date'] == df1['ranking_date'].min()]

            dicionario['likes'] = obs_last_reg['likes'].iloc[0]
            dicionario['views'] = obs_last_reg['views'].iloc[0]
            dicionario['comments'] = obs_last_reg['comments'].iloc[0]
            dicionario['trending_start_date'] = obs_fst_reg['ranking_date'].iloc[0]
            dicionario['trending_end_date'] = obs_last_reg['ranking_date'].iloc[0]
        else:
            ## INCLUDING VALUES INTO NEW DATAFRAME
            dicionario['video_name'] = df1['video_name'].iloc[0]
            dicionario['video_channel_name'] = df1['video_channel_name'].iloc[0]
            dicionario['video_url'] = df1['video_url'].iloc[0]
            dicionario['video_channel_account'] = df1['video_channel_account'].iloc[0]
            dicionario['keywords'] = df1['keywords'].iloc[0]
            dicionario['rank_trend'] = df1['rank_trend'].iloc[0]
            dicionario['likes'] = df1['likes'].iloc[0]
            dicionario['views'] = df1['views'].iloc[0]
            dicionario['comments'] = df1['comments'].iloc[0]
            dicionario['trending_start_date'] = df1['ranking_date'].iloc[0]
            dicionario['trending_end_date'] = df1['ranking_date'].iloc[0]
        
        df_final.loc[len(df_final.index)] = dicionario

    return df_final
            
```

```python
## RUN FUNCTION
df_trending_final = unifyVideos(df_temp2)
df_trending_final.head(10)
```

ADJUSTING OUTPUT FILES IN WEB CRAWLER 

```python
### adjusting output files
import os, json
import glob
from selenium import webdriver
from time import sleep
from scrapy.selector import Selector
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
## READ JSON FILE
path_to_json = '../youtube_trending/output2/' #JSONS PATH
base_url = 'https://www.youtube.com/'

json_pattern = os.path.join(path_to_json,'*.json')
file_list = glob.glob(json_pattern)

## OPEN WEB DRIVER
driver = webdriver.Firefox()

for file in file_list:
    data = pd.read_json(file) # READ DATAFRAME
    print ('Read file: ' + file)
    
    accounts = []
    keywords = []

    ## GET LINKS
    for link in data['video_url']:
        driver.get(link[0])
        sleep(5)
        sel = Selector(text=driver.page_source)
        
        ac = sel.xpath('//ytd-video-owner-renderer/a/@href').extract_first()
        ac = ac.split('/')[1] if ac else None
        accounts.append(ac)

        if ac:
            driver.get(base_url + ac)
            sleep(5)
            sel = Selector(text=driver.page_source)
    
            kw = sel.xpath('//meta[@property="og:video:tag"]/@content').extract()
            keywords.append(kw)
        else:
            keywords.append(None)

    data['video_channel_account'] = accounts
    data['keywords'] = keywords

    ## SAVE DATAFRAME TO JSON
    data.to_json(file, orient="records")
    sleep(3)
    print ('Finish file: ' + file)
    print ("####################################################################")
        
# data['video_url'][0][0]
```
