```python
### adjusting output files
import os, json
import glob
from selenium import webdriver
from time import sleep
from scrapy.selector import Selector
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
## READ JSON FILE
path_to_json = '../youtube_trending/output2/' #JSONS PATH
base_url = 'https://www.youtube.com/'

json_pattern = os.path.join(path_to_json,'*.json')
file_list = glob.glob(json_pattern)

## OPEN WEB DRIVER
driver = webdriver.Firefox()

for file in file_list:
    data = pd.read_json(file) # READ DATAFRAME
    print ('Read file: ' + file)
    
    accounts = []
    keywords = []

    ## GET LINKS
    for link in data['video_url']:
        driver.get(link[0])
        sleep(5)
        sel = Selector(text=driver.page_source)
        
        ac = sel.xpath('//ytd-video-owner-renderer/a/@href').extract_first()
        ac = ac.split('/')[1] if ac else None
        accounts.append(ac)

        if ac:
            driver.get(base_url + ac)
            sleep(5)
            sel = Selector(text=driver.page_source)
    
            kw = sel.xpath('//meta[@property="og:video:tag"]/@content').extract()
            keywords.append(kw)
        else:
            keywords.append(None)

    data['video_channel_account'] = accounts
    data['keywords'] = keywords

    ## SAVE DATAFRAME TO JSON
    data.to_json(file)
    sleep(3)
    print ('Finish file: ' + file)
    print ("####################################################################")
        
# data['video_url'][0][0]
```
