```python
##IMPORTS
import pandas as pd
import numpy as np
```

```python
## READ CSV FILE

df1 = pd.read_csv("../datasets/top1000_2022/social media influencers-youtube - --nov 2022.csv", encoding='utf-8')
df1.head()
```

```python
### FUNCTION TO CONVERT STRING MILLION/THOUSAND FORMAT TO NUMBER

def stringToNumeric(text, bad_data_val = 0):
    d = {
        'K': 1000,
        'M': 1000000,
        'B': 1000000000
    }
    if not isinstance(text, str):
        # Non-strings are bad are missing data in submission
        return bad_data_val

    elif text[-1] in d:
        # separate out the K, M, or B
        num, magnitude = text[:-1], text[-1]
        return int(float(num) * d[magnitude])
    else:
        return float(text)
```

```python
# GET COLUMNS NAME
print(df1.columns.values)
```

```python
## RENAME COLUMNS
df1.rename(columns={'Youtube channel': 'nome_canal',
                    'youtuber name': 'perfil_canal',
                    'Country': 'pais',
                    'Followers': 'inscritos',
                    '\nViews (Avg.)': 'avg_views',
                    'Likes (Avg.)': 'avg_likes',
                    'Comments (Avg.)': 'avg_comments',
                   }, inplace=True)
```

```python
## REMOVE UNUSED COLUMNS
df1.drop(columns=['s.no', 'Category', 'Category-2'], inplace=True)
```

```python
## CHANGE STRING VALUES INTO AVG VALUES (SUBSCRIBERS, VIEWS, LIKES AND COMMENTS) TO FLOAT
df2 = df1

df2["inscritos"] = df2['inscritos'].apply(lambda x: stringToNumeric(x))
df2["avg_views"] = df2['avg_views'].apply(lambda x: stringToNumeric(x))
df2["avg_likes"] = df2['avg_likes'].apply(lambda x: stringToNumeric(x))
df2["avg_comments"] = df2['avg_comments'].apply(lambda x: stringToNumeric(x))
df2.head()
```

```python
## SAVE DATAFRAME INTO CSV FILE

df2.to_csv('../datasets/top1000_2022_cleaned/youtube-nov2022.csv', index=False)
```
